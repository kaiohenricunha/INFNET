```python
# Obtenha, usando requests ou urllib, dentro de seu programa em Python, o csv do link:
# https://sites.google.com/site/dr2fundamentospython/arquivos/Winter_Olympics_Medals.csv e:
# a) Dentre os seguintes países nórdicos: Suécia, Dinamarca e Noruega, verifique: No século XXI (a partir de 2001), qual foi o maior medalhista de ouro, considerando apenas as seguintes modalidades:
#     i. Curling
#     ii. Patinação no gelo (skating)
#     iii. Esqui (skiing)
#     iv. Hóquei sobre o gelo (ice hockey)
# b) Para cada esporte, considere todas as modalidades, tanto no masculino quanto no feminino. Sua resposta deve imprimir um relatório mostrando o total de medalhas de cada um dos países e em que esporte, ano, cidade e gênero (masculino ou feminino) cada medalha foi obtida.

import urllib
import requests
import csv

url = 'https://sites.google.com/site/dr2fundamentospython/arquivos/Winter_Olympics_Medals.csv'
csv = requests.get(url).text
lines = csv.splitlines()

SWE_Gold = 0
NOR_Gold = 0
DEN_Gold = 0

for line in lines:
    if ("SWE" in line) or ("NOR" in line) or ("DEN" in line):
        if ("2006" in line) or ("2005" in line) or ("2004" in line) or ("2003" in line) or ("2002" in line) or ("2001" in line):
            if ("Curling" in line) or ("Skating" in line) or ("Skiing" in line) or ("Ice Hockey" in line):
                if ("Gold" in line):
                    if "SWE" in line:
                        SWE_Gold += 1
                    if "NOR" in line:
                        NOR_Gold += 1
                    if "DEN" in line:
                        DEN_Gold += 1

print("\na) Dentre os seguintes países nórdicos: Suécia, Dinamarca e Noruega, verifique: No século XXI (a partir de 2001), qual foi o maior medalhista de ouro, considerando apenas as seguintes modalidades: \n")

print(f"==> O numero de medalhas de ouro da Suecia foi {SWE_Gold}.")
print(f"==> O numero de medalhas de ouro da Noruega foi {NOR_Gold}.")
print(f"==> O numero de medalhas de ouro da Dinamarca foi {DEN_Gold}.")
print(f"====> O maior campeão de medalhas do OURO foi a NORUEGA com {NOR_Gold} de ouro.\n")

print("\nb) Para cada esporte, considere todas as modalidades, tanto no masculino quanto no feminino. Sua resposta deve imprimir um relatório mostrando o total de medalhas de cada um dos países e em que esporte, ano, cidade e gênero (masculino ou feminino) cada medalha foi obtida.\n")

count = 0
for line in lines:
    colunas = lines[count].split(',')
    

    if colunas[6] == 'Event gender': 
        colunas[6] = 'Gênero'
    elif colunas[6] == 'M': 
        colunas[6] = 'Masculino'
    elif colunas[6] == 'W':
        colunas[6] = 'Feminino'
    elif colunas[6] == 'X':
        colunas[6] = 'Ambos' 

    if colunas[4] == 'NOC':
        colunas[4] = 'País'
    if colunas[2] == 'Sport':
        colunas[2] = 'Esporte'
    if colunas[0] == 'Year':
        colunas[0] = 'Ano'
    if colunas[1] == 'City':
        colunas[1] = 'Cidade'
    if colunas[7] == 'Medal':
        colunas[7] = 'Medalha'

    count += 1 
    tabela_formatada = (f"|{colunas[4]:^6}|{colunas[7]:^9}|{colunas[2]:^10}|{colunas[0]:^6}|{colunas[1]:^23}|{colunas[6]:^14}|")
    print(tabela_formatada)
    
```

```python
import requests
from collections import Counter

url =' https://sites.google.com/site/dr2fundamentospython/arquivos/Video_Games_Sales_as_at_22_Dec_2016.csv'
csv = requests.get(url).text
lines = csv.splitlines()
# lines_pre_format = str(lines).replace(',,,,,,','').replace(',,,,','').replace(',,,','').replace(',,','')


print("8. Obtenha, usando requests ou urllib, dentro de seu programa em Python, o csv do link: https://sites.google.com/site/dr2fundamentospython/arquivos/Video_Games_Sales_as_at_22_Dec_2016.csv. \nObtenha, dentre os jogos do gênero de ação (Action), tiro (Shooter) e plataforma (Platform):")

table_games = []
count = 0
for line in lines:
    colunas = lines[count].split(',')
    table_games.append(colunas)
    count+=1

#Aqui vai criar uma tabela com todas as empresas e seus atributos (como vendas, ano, etc) que fabricaram somente nesse estilo de jogo
table_games_a_s_p = []
for line in table_games:
    if ("Action" in line) or ("Shooter" in line) or ("Platform" in line):
        table_games_a_s_p.append(line)

# Aqui vai criar uma tabela somentes com os nomes das EMPRESAS que fabricaram nesses estilos para que eu possa usar o Count
count = 0
marcas = []
for i in table_games_a_s_p:
    marcas.append(table_games_a_s_p[count][4])
    count+=1

contador_de_jogos = Counter(marcas)

# mostro toda a biblioteca counter, o primeiro é sempre o que tem mais
# print(contador_de_jogos)
print("\na) Quais são as três marcas que mais publicaram jogos dos três gêneros combinados? Indique também o total de jogos de cada marca.\n")

print(f"As três Marcas que mais publicaram jogos foram:\n"
    f" 1 - Activision com {contador_de_jogos['Activision']} jogos publicados\n"
    f" 2 - Ubisoft com {contador_de_jogos['Ubisoft']} jogos publicados\n"
    f" 3 - Electronic Arts com {contador_de_jogos['Electronic Arts']} jogos publicados\n"
    )

#Aqu ieu crio listas das empresas e da quantidade de vendas globais delas
count = 0
vendas_globais = []
empresas = []
for i in table_games_a_s_p:
    empresas.append(table_games_a_s_p[count][4])
    vendas_globais.append(table_games_a_s_p[count][9])
    count+=1

#Aqui eu transformo num dicionario, juntando as lsitas das empresas e das vendas globais
#Importante saber que o dicionario vai substituindo as empresas que são iguais
dic_vendas_empresa = {}
count = 0
for i in range(len(empresas)):   
    if empresas[count] not in dic_vendas_empresa:
        dic_vendas_empresa[empresas[count]] = str(vendas_globais[count]  )
    count+=1

# Mostra todo o dicionario, do maior pro menor
# print(dic_vendas_empresa)
print("b) Quais são as três marcas que mais venderam os três gêneros combinados? Indique também o total de vendas de cada marca.\n")

print(f"As três Marcas que mais venderam os gêneros combinados foram:\n"
    f" 1 - Nintendo com {dic_vendas_empresa['Nintendo']} u.m. jogos vendidos\n"
    f" 2 - Take-Two Interactive com {dic_vendas_empresa['Take-Two Interactive']} u.m. jogos vendidos\n"
    f" 3 - Activision com {dic_vendas_empresa['Activision']} u.m. jogos vendidos\n"
    )

#Aqui eu percoro todos os anos para saber qual o maior/ultimo ano registrado (2017)
years = []
count = 0
for i in table_games_a_s_p:
    if table_games_a_s_p[count][2].isdigit():
        years.append(table_games_a_s_p[count][2])
    count += 1

current_year = max(years)

# Aqui eu cria uma lsita com o numero de recorrencias dos jogos e a quantidade de vendas (Vou precisar das vendas na letra C), primeiro eu verifico se dentro da lista dos jogos, na coluna dos anos, tudo é realmente um digito/numero, porque tem algumas sujeiras que não são (Como por exemplo cabeçalho), se forem ai faço outra condição. A condição é se dentro da lista os anos são maiores que 2007, que é o ultimo ano registrado MENOS 10 (A Questão pede os ultimos 10 anos)
current_games = []
current_sales = []
current_publisher = []
count = 0
for i in table_games_a_s_p:
    if table_games_a_s_p[count][2].isdigit():
        if int(table_games_a_s_p[count][2]) > (int(current_year) - 10):
            current_games.append(table_games_a_s_p[count][0])
            current_sales.append(table_games_a_s_p[count][9])
            current_publisher.append(table_games_a_s_p[count][4])
    count += 1

counter_games = Counter(current_publisher)

print( "c) Qual é a marca com mais publicações em cada um dos gêneros nos últimos dez anos no Japão? Indique também o número total de jogos dela.")
print(f"\nA Marca com maiores publicações nestes TRÊS gêneros nos ultimos 10 anos foi:\n"
    f" 1 - Activision com {counter_games['Activision']} jogos publicados\n"
)

dic_vendas_current_publisher = {}
count = 0
for i in range(len(current_publisher)):   
    if current_publisher[count] not in dic_vendas_current_publisher:
        dic_vendas_current_publisher[current_publisher[count]] = str(current_sales[count])
    count+=1

print("d) Qual foi a marca que mais vendeu em cada um desses gêneros nos últimos dez anos, no Japão? Indique também o total de vendas dela.\n")

print(f"1 - A Nintendo com {dic_vendas_current_publisher['Nintendo']} u.m. jogos vendidos nos ultimos 10 anos.\n")




```
